```python
import mglearn
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import math
```

```python
file= r'C:\Users\Pratik\Desktop\FA\ML\weight_height.csv'

df=pd.read_csv(file)
```

```python
df.head()
```

```python
X = df.drop(columns = 'Weight', axis = 1)
y = df['Weight'].values

plt.scatter(X,y, s = 1)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3, random_state = 42)
```

```python
lr = LinearRegression().fit(X_train, y_train)
```

```python
print("training set score: %f" % lr.score(X_train, y_train))
print("test set score: %f" % lr.score(X_test, y_test))
```

```python
print("lr.coef_: %s" % lr.coef_)
print("lr.intercept_: %s" % lr.intercept_)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print("Mean squared error (MSE): %.2f" % mean_squared_error(y_test, y_pred))
print("Root Mean squared error (RMSE): %.2f" % math.sqrt(mean_squared_error(y_test, y_pred)))
print('Variance score: %.5f' % r2_score(y_test, y_pred))
```

```python
plt.scatter(X,y, s = 1)
plt.plot(X_test, y_pred, color = "red", linewidth = 2)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

```python
lr_final = LinearRegression().fit(X, y)
y_final = lr_final.predict(X)

plt.scatter(X,y, s = 1)
plt.plot(X, y_final, color = "red", linewidth = 2)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()

print("Coefficient = ", lr_final.coef_)
print("Intercept = ", lr_final.intercept_)
```
