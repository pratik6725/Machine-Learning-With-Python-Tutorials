```python
#Binning
df_bins = one_hot_encoded_training_predictors.copy()
print(df_bins['Income'].unique())
df_bins['Income'] =  df_bins['Income'].replace([0,1], 0)
df_bins['Income'] =  df_bins['Income'].replace([2,3], 1)
print(df_bins['Income'].unique())
```

```python
X = df_bins.drop('Income',axis=1)
y = df_bins['Income']
```

```python
#X = df.drop('Income',axis=1)
#y = df['Income']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25,random_state=0)

```

```python
#KNN
clf = KNeighborsClassifier(n_neighbors=3)

clf.fit(X_train, y_train)

training_accuracy = clf.score(X_train, y_train)
test_accuracy = clf.score(X_test, y_test)
print("KNN Classifier")
print("Train : %f" %  training_accuracy)
print("Test : %f" % test_accuracy )
```

```python
#RANDOM FOREST
clf = RandomForestClassifier(n_jobs=-1)
clf.fit(X_train, y_train)
training_accuracy = clf.score(X_train, y_train)
test_accuracy = clf.score(X_test, y_test)
print("Random Forest")
print("Train : %f" %  training_accuracy)
print("Test : %f" % test_accuracy )
```

```python
from sklearn.tree import DecisionTreeClassifier
best_score = 0

for depth in range(2,10):
    for split in range(2,5):
        for leaf in range(5,20):
            for criteria in ['gini', 'entropy']:
                tree_grid_search = DecisionTreeClassifier(criterion = criteria,
                                                         max_depth = depth,
                                                         min_samples_split = split,
                                                         max_leaf_nodes = leaf)
                tree_grid_search.fit(X_train, y_train)
                
                # evaluate the score on test
                score = tree_grid_search.score(X_test, y_test)
                
                if score > best_score:
                    best_score = score
                    best_parameters = {'Depth': depth,
                                      'Split': split,
                                      'Leaves': leaf,
                                      'Criteria' : criteria}

print("Best score = ", best_score)
print("Best Parameters = ", best_parameters)
```

```python
from sklearn.model_selection import GridSearchCV
depth = list(range(2,10))
split = list(range(2,5))
leaf = list(range(5,20))
criteria = ['gini','entropy']

param_grid = { 'max_depth' : depth,
             'min_samples_split' : split,
             'max_leaf_nodes' : leaf,
             'criterion' : criteria}



tree_grid_cv = DecisionTreeClassifier()

grid_search = GridSearchCV(tree_grid_cv, param_grid, cv=5)

grid_search.fit(X_train, y_train)
```

```python
print("Test Accuracy score = ", grid_search.score(X_test, y_test))
```

```python
print("Best Parameters = ", grid_search.best_params_)
print("Best score = ", grid_search.best_score_)
```
